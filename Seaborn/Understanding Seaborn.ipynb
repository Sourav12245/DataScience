```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
test = pd.read_csv('Iris.csv')
test.head()
```

```python
test['species_number'] = ''
for i in range(0,len(test.Species)):
    if test['Species'].iloc[i] == 'Iris-setosa':
        test['species_number'].iloc[i] = 1
    elif test['Species'].iloc[i] == 'Iris-versicolor':
        test['species_number'].iloc[i] = 2
    else:
        test['species_number'].iloc[i] = 3
```

```python
test.head()
```

```python
import warnings
warnings.filterwarnings('ignore')
test = test.drop('Id',axis=1)
```

## DistPlot

```python
sns.distplot(test.SepalLengthCm,kde=True,hist=False)
```

```python
from scipy import stats
sns.distplot(test.PetalLengthCm, fit=stats.gamma, kde = False, hist = False)
```

## KDE Plot

```python
sns.kdeplot(test.SepalLengthCm)
```

```python
sns.kdeplot(test.PetalLengthCm, test.PetalWidthCm,shade=False)
```

```python
sns.kdeplot(test.SepalWidthCm, cumulative=True, clip=(2.5,4.0))
```

## PairPlot

```python
sns.pairplot(test, hue='Species',markers=['o','s','D'])
```

```python
sns.pairplot(test, vars=['SepalLengthCm','PetalLengthCm'],kind='reg')
```

```python
sns.pairplot(test,x_vars=['SepalLengthCm','PetalLengthCm'],y_vars=['SepalWidthCm','PetalWidthCm'],hue= 'Species')
```

## Stripplot and swampplot
##### Scatter plots better be used for small set of data. Large data sets will freek out the computer

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.stripplot(data=tips,x='day',y='tip',jitter=True, hue='size',split=True)
```

```python
sns.boxplot(data=tips, x='day',y='tip')
sns.swarmplot(data=tips, x='day',y='tip',palette='Set1')
```

## Box Plot

```python
sns.boxplot(data=tips, x='total_bill')
```

```python
sns.boxplot(data=tips, x='total_bill',notch=True)
```

```python
sns.boxplot(data=tips, x='total_bill',y='day')
```

```python
sns.boxplot(data=tips, y='total_bill',x='day')
sns.swarmplot(data=tips, y='total_bill',x='day',color='Black')
```

## JointPlot

```python
sns.jointplot(data=tips,x='total_bill',y='tip')
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='reg')
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip').plot_joint(sns.kdeplot)
```

## Violinplot

```python
sns.violinplot(data=tips,x='total_bill')
```

```python
sns.violinplot(data=tips,x='total_bill',inner='box')
```
